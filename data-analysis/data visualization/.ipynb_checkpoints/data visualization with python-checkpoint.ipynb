```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,3]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(['This is y', 'This is z'])
plt.show()
```

```python
data = pd.read_csv('./data/countries.csv')
data.head()
```

```python
us
```

```python
china = data[data.country == 'China']
```
